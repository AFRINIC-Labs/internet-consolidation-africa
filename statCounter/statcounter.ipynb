```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

markers = [",","v","o","p",".","s","+","D","1","2","3","4","P","h","H"]

```

## Social media statistics (2009 - 2019)
https://gs.statcounter.com/social-media-stats/all/africa#monthly-200901-201908

```python
df_sm = pd.read_csv('data/social_media-af-monthly-200901-201908.csv', parse_dates=['Date'], index_col='Date')
df_sm.sort_index(ascending=True,inplace=True)
df_sm.head()
```

```python
def plotname(df, name, marker):
    data = df[name]
    plt.plot(data.index, data.values, marker=marker)
```

```python
plt.figure(figsize = (12, 5.5))

names = df_sm.columns.values
print(names)
i=0
for name in names:
    plotname(df_sm, name, markers[i])
    i=i+1

plt.legend(names)
plt.grid()
plt.title('Social media popularity 2009-2019')
plt.ylabel('Percentage')
plt.xlabel('Year')
plt.savefig('plots/social_media_popularity.png')
plt.show()
```

## Browser market share in Africa (2009 - 2019)
http://gs.statcounter.com/browser-market-share/all/africa#monthly-200901-201908

```python
df_bw = pd.read_csv('data/browser-af-monthly-200901-201908.csv', parse_dates=['Date'], index_col='Date')
df_bw.sort_index(ascending=True,inplace=True)
df_bw.head()
```

```python
plt.figure(figsize = (12, 5.5))

names = df_bw.columns.values
i=0
for name in names:
    try:
        plotname(df_bw, name, markers[i])
    except IndexError as e:
        plotname(df_bw, name, '')
    i=i+1

plt.legend(names[:6])
plt.grid()
plt.title('Browser popularity 2009-2019')
plt.ylabel('Percentage')
plt.xlabel('Year')
plt.savefig('plots/browser_popularity.png')
plt.show()
```
